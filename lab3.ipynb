```python
amount = True
count = 0
count_pass = 0
while(amount):
    print("Options Available")
    print("A = Add a new grade")
    print("R = Report")
    print("Q = Quit")
    user_input = input("Select an Option")
    if(user_input == "A"):
        new_grade = int(input("Enter a grade"))
        count = count + 1
        if(new_grade < 50):
            print("Fail")
        elif(new_grade >= 50):
            print("Pass")
            count_pass = count_pass + 1
    elif(user_input == "R"):
        print("Report")
        print("number of grades entered: " + str(count))
        print("number of passing grades entered: " + str(count_pass))
    elif(user_input == "Q"):
        amount = False
        print("System Closed")
    else:
        print("invalid entry, please try again.")
    
```
